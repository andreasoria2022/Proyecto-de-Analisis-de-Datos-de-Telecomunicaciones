```python
df_internet_acceso_tec['Cablemodem'] = df_internet_acceso_tec['Cablemodem'].astype(int)
```

```python
df_internet_acceso_tec['Fibra óptica'] = df_internet_acceso_tec['Fibra óptica'].astype(int)
```

```python
df_internet_acceso_tec['Otros'] = df_internet_acceso_tec['Otros'].astype(int)
```

```python
df_internet_acceso_tec['Total'] = df_internet_acceso_tec['Total'].astype(int)
```

Reemplazo en la columna año '2019 *' por '2019'

```python
df_internet_acceso_tec['Año']=df_internet_acceso_tec['Año'].replace('2019 *', '2019')
```

Reemplazo en la columna Trimestre '3 *' por '3'

Una vez que realice las modificaciones que considere necesarias. Crearé un csv para guardar el dataframe df_internet_acceso_tec.

Voy a crear un dataframe a partir de un csv

Elimino la columna Unnamed: 13

Realizare los cambios de tipo de datos de object a float.

```python
df_acceso_por_tec_y_localidad['ADSL']= df_acceso_por_tec_y_localidad['ADSL'].replace ('- 0','0').astype(str).str.replace('.', '').str.replace(',', '.').astype(float)
df_acceso_por_tec_y_localidad['CABLEMODEM']= df_acceso_por_tec_y_localidad['CABLEMODEM'].replace ('- 0','0').astype(str).str.replace('.', '').str.replace(',', '.').astype(float)
df_acceso_por_tec_y_localidad['DIAL UP']= df_acceso_por_tec_y_localidad['DIAL UP'].replace ('- 0','0').astype(str).str.replace('.', '').str.replace(',', '.').astype(float)
df_acceso_por_tec_y_localidad['FIBRA OPTICA']= df_acceso_por_tec_y_localidad['FIBRA OPTICA'].replace ('- 0','0').astype(str).str.replace('.', '').str.replace(',', '.').astype(float)
df_acceso_por_tec_y_localidad['OTROS']= df_acceso_por_tec_y_localidad['OTROS'].replace ('- 0','0').astype(str).str.replace('.', '').str.replace(',', '.').astype(float)
df_acceso_por_tec_y_localidad['SATELITAL']= df_acceso_por_tec_y_localidad['SATELITAL'].replace ('- 0','0').astype(str).str.replace('.', '').str.replace(',', '.').astype(float)
df_acceso_por_tec_y_localidad['WIMAX']= df_acceso_por_tec_y_localidad['WIMAX'].replace ('- 0','0').astype(str).str.replace('.', '').str.replace(',', '.').astype(float)
df_acceso_por_tec_y_localidad['WIRELESS']= df_acceso_por_tec_y_localidad['WIRELESS'].replace ('- 0','0').astype(str).str.replace('.', '').str.replace(',', '.').astype(float)
df_acceso_por_tec_y_localidad['Total general']= df_acceso_por_tec_y_localidad['Total general'].replace ('- 0','0').astype(str).str.replace('.', '').str.replace(',', '.').astype(float)
```
